```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("heart_failure_clinical_records_dataset.csv")
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

sns.distplot(data['DEATH_EVENT'], kde=True)
plt.title('Distribution of DEATH_EVENT')
plt.show()
```

```python
for column in data.columns:
    sns.displot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()

fig = px.scatter(data, x='DEATH_EVENT', y='smoking', title='Scatter Plot - smoking vs ejection_fraction')
fig.show()
```

```python
plt.figure(figsize =(10,8))
sns.barplot(y= "DEATH_EVENT", x="smoking", data = data)
sns.despine(bottom =True, left =True)
plt.title("smoking of difference DEATH_EVENT")
```

```python
fig = px.violin(data, y="DEATH_EVENT", x="smoking", box=True, points="all", hover_data=data.columns)
fig.show()
```

```python
sns.pairplot(data)
```

```python
import matplotlib.pyplot as plt

correlation_matrix = data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
